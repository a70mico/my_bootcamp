```python
# primera prueba de hipotesis

#una fabrica produce 150 unidades/hora con una desviacion estandar de 8 unidades, una maquina nueva produce 160 unidades/hora con una desviacion estandar de 8 unidades. total de muestras n = 35 
# el nivel de signifcancia lo establecemos alfa = 0.01 porque la maquina es cara y queremos ser cautos.

z_score = (160 - 150) / (8/np.sqrt(35))
print(z_score)
p_value = 2*(1-stats.norm.cdf(abs(z_score)))
print(p_value)
```

```python
#un automovil a 30 millas/h debe recorrer max 65 pies al frenar, un fabricante de automoviles 
# dise;o un nuevo sistema de frenos que recorre una distancia de 63.5 pies con una desviacion estandar de 4 pies habiendolo probado en 64 autos
# alfa = 0.01 por seguridad de ser humanos

z_score = (63.5 - 65) / (4/np.sqrt(64))
print(z_score)
p_value = (1-stats.norm.cdf(abs(z_score)))
print(p_value)
```
